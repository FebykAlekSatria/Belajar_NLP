```python
from nltk.stem.porter import PorterStemmer
from nltk.stem.lancaster import LancasterStemmer
from nltk.stem.snowball import SnowballStemmer

from nltk.stem import WordNetLemmatizer
import pandas as pd
```

```python
inputWords = ['writing', 'wrote', 'written', 'friends', 'friendship', 'connect',
 'connected', 'connections', 'randomize', 'caring', 'played', 'higher']

porter = PorterStemmer()
lancaster = LancasterStemmer()
snowball = SnowballStemmer('english')

wordNetLemmatizer = WordNetLemmatizer()
```

```python
porterResult = [porter.stem(w) for w in inputWords]
lancasterResult = [lancaster.stem(w) for w in inputWords]
snowballResult = [snowball.stem(w) for w in inputWords]
lematizerResultV = [wordNetLemmatizer.lemmatize(w, pos='v') for w in inputWords]
lematizerResultN = [wordNetLemmatizer.lemmatize(w, pos='n') for w in inputWords]
```

```python
pd.DataFrame({
    'inputWords' : inputWords,
    'porter' : porterResult,
    'lancaster' : lancasterResult,
    'snowball' : snowballResult,
    'lematizerV' : lematizerResultV,
    'lematizerN' : lematizerResultN
})
```
